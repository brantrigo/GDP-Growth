GDP (constant LCU) Local Currency fixes a base year and measure the increment. The problem is as further you go from that year, the less meaningful the figures becomes.

GDP Growth is in respect to last year in constant currency. Both GDP Growth and GDP Constant suffers from using population size. 

GDP Growth per capita conglomerates all.

GDP per capita, PPP (purchasing power parity) takes into considaration prices, life level.

```python
# Indicators that contain GDP within their description name
pd.read_sql("""SELECT IndicatorCode, IndicatorName 
FROM Indicators WHERE IndicatorName LIKE '%GDP%';""", conn)
```

Create a dataframe for the UUSS showing all Indicator Codes, their values and years

```python
usa_data = pd.read_sql("""SELECT IndicatorName, Value, Year
    FROM Indicators LEFT JOIN CountryIndicators
    ON Indicators.IndicatorCode = CountryIndicators.IndicatorCode
    WHERE
    CountryIndicators.CountryCode = "USA";""", conn)
# CountryIndicators.Year = "1960" AND
```

```python
usa_data
df = usa_data.pivot(index='Year', columns='IndicatorName', values='Value')
df[0:10]
```

Select only Indicators that have all values available 

```python
# Set the threshold to n-1 = 50, since some important indicators do
# not have values for first year
df2 = df.dropna(axis=1, how="any", thresh=50, subset=None, inplace=False)
df2
```

```python
series = df2.loc["1961":"2010", "Total reserves minus gold (current US$)"]
pd.plotting.autocorrelation_plot(series)
#  GDP growth (annual %)
```

```python
i = 0
series = df2.iloc[1:49, i]
pd.plotting.autocorrelation_plot(series)
```
